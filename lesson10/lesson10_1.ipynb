```python
import random

file = open("names.txt",encoding='utf-8')
contents:list[str] = file.readlines()
file.close()
names:list[str] = [name.strip() for name in contents]
pick = int(input("請輸入所需要的姓名數量:"));

for idx, name in enumerate(random.sample(names, pick)):
    print(f'{idx+1}.{name}')
```

```python
import random

file = open("names.txt",encoding='utf-8')
contents:list[str] = file.readlines()
file.close()
names:list[str] = [name.strip() for name in contents]
pick = int(input("請輸入所需要的姓名數量:"));
student_1 = random.sample(names, pick)
student = [
    {
        '姓名':name,
        '國文':random.randint(50,100),
        '英文':random.randint(50,100),
        '數學':random.randint(50, 100) 
    } 
    for name in student_1]
print(student)
```

```python
import csv

filednames = student[0].keys()

with open('students.csv','w',encoding='utf-8') as csvfile:
    writer = csv.DictWriter(csvfile,fieldnames=filednames) #fieldnames欄位名稱
    # 寫入欄位標題
    writer.writeheader()
    
    # 寫入多筆資料
    writer.writerows(student)
print("寫入成功")
```
